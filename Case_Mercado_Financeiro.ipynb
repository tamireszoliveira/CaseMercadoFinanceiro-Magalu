```python
# libs para Modelagem e Matrizes
import numpy as np
import pandas as pd

# libs para análises gráficas
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.graph_objects as go

# lib para ignorar avisos
import warnings

# desabilitando avisos
warnings.filterwarnings('ignore')
```

```python
#lendo a base de dados
df = pd.read_excel('Vase_004 - Magalu - Sem Resolução.xlsx')
#dimensao
df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
# series temporais
Dados = df.set_index('Data')
Dados.head()
```

```python
sns.set_style('dark')
plt.figure(figsize=(16, 5))
plt.title('Análise das Ações da Magalu - Fechamento', fontsize=15, loc='left')
plt.plot(Dados.index, Dados['Fechamento']);

plt.xlabel('Período da Cotação')
plt.ylabel('Valor da Ação(R$)');
```

Valor das ações ao final de 2021:

```python
Dados.tail()
```

```python
Media_Movel = Dados['Fechamento'].rolling(5).mean()
Media_Tendencia = Dados['Fechamento'].rolling(30).mean()

sns.set_style('dark')
plt.figure(figsize=(16, 5))
plt.title('Análise das Ações da Magalu - Fechamento', fontsize=15, loc='left')

plt.plot(Dados.index, Dados['Fechamento']);
plt.plot(Media_Movel.index, Media_Movel)
plt.plot(Media_Tendencia.index, Media_Tendencia)
plt.legend(loc='upper right')


plt.xlabel('Período da Cotação')
plt.ylabel('Valor da Ação(R$)');
```

🟦 Linha de fechamento
🟧 Média móvel de 5 dias
🟩 Média de tendência de 30 dias

```python
# boxplot mensal
df['Mes'] = df['Data'].dt.month
plt.figure(figsize=(16, 5))
sns.boxplot(data = df, x='Mes', y='Fechamento', hue = 'Mes');
```

```python
df.groupby(['Mes']).describe()['Fechamento']
```

```python
Grafico = go.Figure(
    data = [
        go.Candlestick(
            x = Dados.index,
            open = Dados['Abertura'],
            high = Dados['Maior'],
            low = Dados['Menor'],
            close = Dados['Fechamento'],
        )
    ]
)
Grafico.show()
```
